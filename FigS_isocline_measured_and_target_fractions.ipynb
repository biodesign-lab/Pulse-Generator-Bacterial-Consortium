### Figure S...

Here we create Figure S... in the supplement.

```python
import os
print(os.getcwd())
import numpy as np
import seaborn as sns
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
# Load the data from matlab .mat file
measured_fracs = sio.loadmat('measured_fracs_time0.mat')
target_fracs = sio.loadmat('target_fracs.mat')



exper = 2 #isocline experiment
day=0 #day 1 of the isocline experiment
rX_m_d1_isoc = measured_fracs['measured_fracs_time0'][0,:,day, exper]   #strain X
rY_m_d1_isoc = measured_fracs['measured_fracs_time0'][1,:,day, exper]   #strain Y
rZ_m_d1_isoc = measured_fracs['measured_fracs_time0'][2,:,day, exper]   #strain Z
day=1 #day 2 of the fixZ experiment
rX_m_d2_isoc = measured_fracs['measured_fracs_time0'][0,:,day, exper]   #strain X
rY_m_d2_isoc = measured_fracs['measured_fracs_time0'][1,:,day, exper]   #strain Y
rZ_m_d2_isoc = measured_fracs['measured_fracs_time0'][2,:,day, exper]   #strain Z
day=2 #day 3 of the fixZ experiment
rX_m_d3_isoc = measured_fracs['measured_fracs_time0'][0,:,day, exper]   #strain X
rY_m_d3_isoc = measured_fracs['measured_fracs_time0'][1,:,day, exper]   #strain Y
rZ_m_d3_isoc = measured_fracs['measured_fracs_time0'][2,:,day, exper]   #strain Z

rX_pipetted_isoc=target_fracs['isoc_pipetted_fracs'][0,:]
rY_pipetted_isoc=target_fracs['isoc_pipetted_fracs'][1,:]
rZ_pipetted_isoc=target_fracs['isoc_pipetted_fracs'][2,:]

rX_target_isoc=target_fracs['ideal_fracs'][0, :, exper]
rY_target_isoc=target_fracs['ideal_fracs'][1, :, exper]
rZ_target_isoc=target_fracs['ideal_fracs'][2, :, exper]




sns.set()
sns.set(font_scale=1.2)
sns.set_style('ticks')
Fig2, axs = plt.subplots(figsize=(6,4))

print(rX_m_d1_isoc)
print(rX_m_d2_isoc)
print(rX_m_d3_isoc)
# Plot the target fractions, pipetted fractions, and measured fractions
#Plot target fractions
axs.plot(rX_target_isoc,rY_target_isoc, linestyle='none', color='b', marker='o', markersize=8, label='target')   

#Plot pipetted fractions
axs.plot(rX_pipetted_isoc,rY_pipetted_isoc, linestyle='none', color='g', marker='o', markersize=8, label='pipetted')   
 
#Plot measured fractions
axs.plot(rX_m_d1_isoc,rY_m_d1_isoc, linestyle='none', color='gray', marker='o', markersize=8, label='measured (day 1)', alpha=0.8)   
axs.plot(rX_m_d2_isoc,rY_m_d2_isoc, linestyle='none', color='gray', marker='o', markersize=8, label='measured (day 2)', alpha=0.5)   
axs.plot(rX_m_d3_isoc,rY_m_d3_isoc, linestyle='none', color='gray', marker='o', markersize=8, label='measured (day 3)', alpha=0.2)   

plt.legend(fontsize=14, frameon=False, bbox_to_anchor=(0.8,1.0), handletextpad=0.5)

#plt.plot([0, 1], [0, 1], linestyle='--', color='k', lw=1, alpha=0.3)
plt.xticks((0,0.4,0.8),(0,0.4,0.8))
plt.yticks((0,0.3,0.6),(0,0.3,.6))
axs.set_xlabel('$r_X$', fontsize=18)
axs.set_ylabel('$r_Y$', fontsize=18)
axs.set_title('Isocline Experiment', fontsize=20)
axs.set_aspect('equal')
sns.despine()


Fig2.savefig("FigS_isocline_fracs.pdf", bbox_inches='tight', transparent=True)
```
